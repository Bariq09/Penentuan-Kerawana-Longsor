```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

```python
print("#Fuzzifikasi Kemiringan tanah")
print("input:")
print("1.Kemiringan tanah")
print("Datar < 10")
print("Landai 10 - 20")
print("Agak Curam 20 -30")
print("Curam 30 -40")
print("Terjal > 40")
```

```python
land_slide = ctrl.Antecedent(np.arange(0,50,1),'Kemiringan Tanah')
slide_x = np.arange(0,50,1)
```

```python
land_slide["Datar"] = fuzz.trapmf(land_slide.universe,[0,0,5,10])
land_slide["Landai"] = fuzz.trimf(land_slide.universe,[5,10,20])
land_slide["Agak Curam"] = fuzz.trimf(land_slide.universe,[15,20,30])
land_slide["Curam"] = fuzz.trimf(land_slide.universe,[25,30,40])
land_slide["Terjal"] = fuzz.trapmf(land_slide.universe,[35,40,50,50])
```

```python
print("#Plotting Kemiringan Tanah")
```

```python
#Kemiringan Tanah Datar
print('#Kemiringan Tanah Datar')
land_datar = fuzz.trapmf(slide_x,[0,0,5,10])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(slide_x, land_datar, 'blue')
ax0.set_title('Kemiringan Tanah Datar')
```

```python
#Kemiringan Tanah Landai
print('#Kemiringan Tanah Landai')
land_landai = fuzz.trimf(slide_x,[5,10,20])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(slide_x, land_landai, 'orange')
ax0.set_title('Kemiringan Tanah Landai')
```

```python
print('#Kemiringan Tanah Agak Curam')
land_agak_curam = fuzz.trimf(slide_x,[15,20,30])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(slide_x, land_agak_curam, 'red')
ax0.set_title('Kemiringan Tanah Datar')
```

```python
#Kemiringan Tanah Curam
print('#Kemiringan Tanah Curam')
land_curam = fuzz.trimf(slide_x,[25,30,40])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(slide_x, land_curam, 'green')
ax0.set_title('Kemiringan Tanah Datar')
```

```python
#Kemiringan Tanah Terjal
print('#Kemiringan Tanah Terjal')
land_terjal = fuzz.trapmf(slide_x,[35,40,50,50])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(slide_x, land_terjal, 'black')
ax0.set_title('Kemiringan Tanah Datar')
```

```python
land_slide.view()
```

```python
#Fuzzifikasi Tipe Tanah
print('#Fuzzifikasi Tipe Tanah')
print('input:')
print('4.CTipe Tanah')
print('Tidak Peka > 10')
print('Kurang Peka 10 - 20')
print('Agak Peka 20 -30')
print('Peka < 40')
```

```python
#Fuzzifikasi Tingkat Kerawanan Bahaya Longsor
print('Tingkat Kerawanan Bahaya Longsor')
print('input:')
print('6. Tingkat Kerawanan Bahaya Longsor')
print('Tidak Rawan < 3.40')
print('Kerawanan Rendah 3.40 - 4.50')
print('Kerawanan Sedang 4,50 - 5.40')
print('Kerawanan Tinggi 5.40 - 7.00')
print('Sangat Rawan > 7.00')
```

```python
avalanche_danger = ctrl.Antecedent(np.arange(0,10,1.0),'Tingkat Kerawanan Longsor')
danger_x = np.arange(0,10,1.0)
```

```python
avalanche_danger["Tidak Rawan"] = fuzz.trapmf(avalanche_danger.universe,[0,0,3.40,3.40])
avalanche_danger["Kerawanan Rendah"] = fuzz.trimf(avalanche_danger.universe,[3.40,3.40,4.50])
avalanche_danger["Kerawanan Sedang"] = fuzz.trimf(avalanche_danger.universe,[4.50,4.50,5.40])
avalanche_danger["Kerawanan Tinggi"] = fuzz.trimf(avalanche_danger.universe,[5.40,5.40,7.00])
avalanche_danger["Sangat Rawan"] = fuzz.trapmf(avalanche_danger.universe,[7.00,7.00,10,10])
```

```python
print('#Plotting Penggunaan Lahan')
```

```python
#Tidak Rawan Longsor
print('#Tidak Rawan Longsor')
avalanche_no = fuzz.trapmf(danger_x,[0,0,3.40,3.40])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(danger_x, avalanche_no, 'red')
ax0.set_title('Tingkat Rawan Longsor')
```

```python
#Kerawanan Longsor Rendah
print('#Kerawanan Longsor Rendah')
avalanche_low = fuzz.trimf(danger_x,[3.40,3.40,4.50])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(danger_x, avalanche_low, 'blue')
ax0.set_title('Kerawanan Longsor Rendah')
```

```python
#Kerawanan Longsor Sedang
print('#Kerawanan Longsor Sedang')
avalanche_sedang = fuzz.trimf(danger_x,[4.50,4.50,5.40])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(danger_x, avalanche_sedang, 'green')
ax0.set_title('Kerawanan Longsor Sedang')
```

```python
#Kerawanan Longsor Tinggi
print('#Kerawanan Longsor Tinggi')
avalanche_tinggi = fuzz.trimf(danger_x,[5.40,5.40,7.00])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(danger_x, avalanche_tinggi, 'yellow')
ax0.set_title('Kerawanan Longsor Tinggi')
```

```python
#Sangat Rawan
print('#Sangat Rawan')
very_high_avalanche= fuzz.trapmf(danger_x,[7.00,7.00,10,10])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(danger_x, very_high_avalanche, 'orange')
ax0.set_title('Sangat Rawan')
```
