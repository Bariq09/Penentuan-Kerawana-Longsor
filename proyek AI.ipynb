```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

# Input Variabel:
## Fuzzifikasi Kemiringan tanah
### input:
### 1.Kemiringan tanah

Datar < 10

Landai 10 - 20

Agak Curam 20 -30

Curam 30 -40

Terjal > 40

```python
land_slide = ctrl.Antecedent(np.arange(0,50,1),'land_slide')
slide_x = np.arange(0,50,1)
```

```python
land_slide["Datar"] = fuzz.trapmf(land_slide.universe,[0,0,5,10])
land_slide["Landai"] = fuzz.trimf(land_slide.universe,[5,10,20])
land_slide["Agak Curam"] = fuzz.trimf(land_slide.universe,[15,20,30])
land_slide["Curam"] = fuzz.trimf(land_slide.universe,[25,30,40])
land_slide["Terjal"] = fuzz.trapmf(land_slide.universe,[35,40,50,50])
```

## Plotting Kemiringan Tanah

### Kemiringan Tanah Datar

```python
#Kemiringan Tanah Datar
print('#Kemiringan Tanah Datar')
land_datar = fuzz.trapmf(slide_x,[0,0,5,10])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(slide_x, land_datar, 'blue')
ax0.set_title('Datar')
```

### Kemiringan Tanah Landai

```python
#Kemiringan Tanah Landai
print('#Kemiringan Tanah Landai')
land_landai = fuzz.trimf(slide_x,[5,10,20])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(slide_x, land_landai, 'orange')
ax0.set_title('Landai')
```

### Kemiringan Tanah Agak Curam

```python
print('#Kemiringan Tanah Agak Curam')
land_agak_curam = fuzz.trimf(slide_x,[15,20,30])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(slide_x, land_agak_curam, 'red')
ax0.set_title('Agak Curam')
```

### Kemiringan Tanah Curam

```python
#Kemiringan Tanah Curam
print('#Kemiringan Tanah Curam')
land_curam = fuzz.trimf(slide_x,[25,30,40])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(slide_x, land_curam, 'green')
ax0.set_title('Curam')
```

### Kemiringan Tanah Terjal

```python
#Kemiringan Tanah Terjal
print('#Kemiringan Tanah Terjal')
land_terjal = fuzz.trapmf(slide_x,[35,40,50,50])
fig, (ax0) = plt.subplots(nrows=1, figsize=(6,6))
ax0.plot(slide_x, land_terjal, 'black')
ax0.set_title('Terjal')
```
